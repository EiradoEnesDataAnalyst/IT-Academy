1. Contador y organizador de palabras de un texto.  
El cliente qued√≥ satisfecho con el contador de palabras, pero ahora quiere que se puedan leer archivos TXT y que se calcule la frecuencia de cada palabra,  
ordenadas dentro de las entradas habituales del diccionario seg√∫n la letra con la que empiezan. Es decir, las claves deben ir de la A a la Z, y dentro de la A debemos ordenar de la A a la Z.

```python
# Importo la librer√≠a Pandas y archivo

import pandas as pd

archivo_txt = r"Archivos\tu me quieres blanca.txt"

df = pd.read_csv(archivo_txt,sep = "\t",header = None)

df
```

##### Texto en Idioma Castellano #####

Teniendo en cuenta que el texto del archivo est√° en idioma castellano,  
vamos a redactar un c√≥digo que limpie y transforme las palabras de manera que el
resultado del diccionario final, no tenga errores.

[RAE,abecedario espa√±ol](https://www.rae.es/dpd/abecedario)

```python
# C√≥digo de funci√≥n: importo libreria RE para limpiar de forma r√°pida y segura

import re

def creaci√≥n_dict(df):

    '''
    Primera parte:
    Limpio el texto para dejarlo preparado y creo
    una lista con todas las palabras, ordenadas de "a-z".
    '''

    lista_palabras_totales = []
    
    for fila in df.iloc[:,0]:
        fila = fila.lower()
        fila_limpia = re.sub(r"[^a-z√°√©√≠√≥√∫√º√± ]", " ", fila)
        fila_limpia = re.sub(r"\s+", " ", fila_limpia).strip()

        lista_palabras = fila_limpia.split()

        lista_palabras_totales.extend(lista_palabras)


    lista_palabras_totales.sort()


    '''
    Segunda parte:
    Creo una lista "abecedario" para generar las claves del diccionario.
    Luego, mediante un bucle, asigno cada palabra a la letra correspondiente
    e incremento su n√∫mero de repeticiones.
    '''

    
    abecedario = []

    for palabra in lista_palabras_totales:
        if palabra[0] not in abecedario:
            abecedario.append(palabra[0])

    abecedario.sort()
 

    diccionario = {letra : {} for letra in abecedario}


    for palabra in lista_palabras_totales:
        letra = palabra[0]
        if letra in diccionario:
            if palabra in diccionario[letra]:
                diccionario[letra][palabra] += 1
            else:
                diccionario[letra][palabra] = 1

    '''    
    Toma el diccionario de esa letra,lo convi√©rte en lista, 
    ord√©na alfab√©ticamente y vuelve a convertirlo en diccionario.
    '''

    for letra in diccionario:
        diccionario[letra] = dict(sorted(diccionario[letra].items()))

        
    return diccionario
```

```python
### Ejecuci√≥n

creaci√≥n_dict(df)
```

2. Conversi√≥n de tipos de datos.  

El cliente recibe una lista de datos y necesita generar dos listas: la primera contendr√° todos los elementos que se pudieron convertir en float,  
y la segunda contendr√° los elementos que no se pudieron convertir.  
Si quieres, tambi√©n puedo ayudarte a escribir el c√≥digo para hacerlo en Python o en otro lenguaje.

```python
# Funci√≥n: utilizar type para verificar el tipo de datos y una funci√≥n interna dentro de la principal

lista_usuario = ['1.3', 'one', '1e10', 'seven', '3-1/2', ('2', 1, 1.4, 'not-a-number'), [1, 2, 3, '3.4']]


def clasificaci√≥n_datos(lista):
    lista_datos_float = []
    lista_datos_no_convertibles = []

    '''
    La funci√≥n recorre toda la estructura (listas, tuplas, sets, dics),
    extrae los valores internos sin importar la profundidad
    e intenta convertir cada uno a float,
    separando los convertibles de los no.
    '''

    def sub_elementos(elemento):
        if type(elemento) in (list,dict,tuple,set):
            for sub_valor in elemento:
                sub_elementos(sub_valor)
        else:
            try:
                float(elemento)
                lista_datos_float.append(float(elemento))
            except:
                lista_datos_no_convertibles.append(elemento)

    sub_elementos(lista)
                

    return lista_datos_float,lista_datos_no_convertibles

```

```python
x = clasificaci√≥n_datos(lista_usuario)


lista_datos_float,lista_datos_no_convertibles = x

print('lista de datos float:\n',lista_datos_float)
print('lista de datos no convertibles:\n',lista_datos_no_convertibles)

```

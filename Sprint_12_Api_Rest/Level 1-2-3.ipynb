<div style="color:#777777; font-family:monospace; width:50%; margin:0 auto; text-align:left;">
<p>The <strong>GET</strong> method is one of the main HTTP verbs and is used to <strong>retrieve data</strong> from a server.<br><br>
A GET request asks the server to return information without changing anything on it. It is commonly used to access public resources or fetch data from APIs.
<br><br>
Key characteristics:
<ul><li>Sends data through the URL.</li><li>Does not modify server state.</li><li>Safe and idempotent.</li></ul>
</div>

<p style="color:#777777;">
Convert the data into a pandas DataFrame:
<p style="color:#777777;">
-Display the first rows of the DataFrame.


```python
# Dataframe 

url = "https://restcountries.com/v3.1/all?fields=name,capital,region,population"
response = requests.get(url)
data = response.json()

clean_data = []

for country in data:
    clean_data.append({"name": country["name"]["common"],
                       "capital": country["capital"][0] if "capital" in country and country["capital"] else None,
                       "region": country.get("region", None),
                       "population": country.get("population", None)})

df = pd.DataFrame(clean_data)

df.head(10)
```

<div style="text-align:center;"><h4 style="color:#F7FF56;">Package Search:</h4></div>
<div style="color:#777777; font-family:monospace; width:50%; margin:0 auto; text-align:left;">
<p>A <strong>package search</strong> refers to an API endpoint or service that allows clients to <strong>look up packages</strong> by name, ID, category, or keywords.<br><br>
It is commonly used in systems that manage software packages, shipments, products, or library modules.
<br><br>
Typical operations:
<ul><li>Search by package name or keyword</li><li>Filter by version, category, or status</li><li>Retrieve package details from a registry or database</li></ul>
<br>
Example endpoint:
<strong>/packages/search?query=express</strong>
</p>
</div>

Using the resource_id of the selected resource, make a query with datastore_search to:
Retrieve at least 100 records.


```python
# Resource_id 

resource_id = "add38a29-8412-4bf2-b7f5-2a7514a9d380"
data_url = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/datastore_search"

params = {"resource_id": resource_id,"limit": 211}

response = requests.get(data_url, params=params).json()

records = response["result"]["records"]
records
```
